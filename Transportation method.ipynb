```python
from collections import defaultdict
import numpy as np

c_d = {}
def change(ch):
    if sum_s>sum_d:
        ch[chr(n+65)] = 0
        return ch
    else:
        return ch

m  = int(input("Enter the number of rows: "))
n  = int(input("Enter the number of rows: "))

sum_s,sum_d = 0,0
demand = {}
print("Input the demand: ")
for i in range(m):
    dem = int(input())
    sum_d = sum_d + dem
    demand[chr(i+65)] = dem
print(demand)

supply = {}
print("Input the supply: ")
for i in range(n):
    sup = int(input())
    sum_s = sum_s + sup
    supply[chr(i+97)] = sup
print(supply)

print("Enter the costs row wise: ")
for i in range(m):
    c_s = {}
    t = 0
    for j in range(n):
        cost = int(input())
        c_s[chr(j+65)] = cost
        if t<n-1:
            c_d[chr(i+97)] = c_s
        else:
            c_d[chr(i+97)] = change(c_s)
        t=t+1
print(c_d)
        
if sum_d>sum_s:
    c_s = {}
    for j in range(n):
        c_s[chr(j+65)] = 0
        c_d[chr(m+97)] = c_s
    supply[chr(n+97)] = sum_d-sum_s

if sum_s>sum_d:
    demand[chr(m+65)] = sum_s-sum_d

cols = sorted(demand.keys())
    
res = dict((k, defaultdict(int)) for k in c_d)
g = {}

#g contains all the columns and rows names like - a,b,c,... and so on
print("Solution by VAM: ")
for x in supply:
    g[x] = sorted(c_d[x].keys(), key=lambda g: c_d[x][g])
for x in demand:
    g[x] = sorted(c_d.keys(), key=lambda g: c_d[g][x])

while g:
    d = {}
    for x in demand:
        d[x] = (c_d[g[x][1]][x] - c_d[g[x][0]][x]) if len(g[x]) > 1 else c_d[g[x][0]][x]
    s = {}
    for x in supply:
        s[x] = (c_d[x][g[x][1]] - c_d[x][g[x][0]]) if len(g[x]) > 1 else c_d[x][g[x][0]]
    f = max(d, key=lambda n: d[n])
    t = max(s, key=lambda n: s[n])
    t, f = (f, g[f][0]) if d[f] > s[t] else (g[t][0], t)
    v = min(supply[f], demand[t])
    res[f][t] += v
    demand[t] -= v
    if demand[t] == 0:
        for k, n in supply.items():
            if n != 0:
                g[k].remove(t)
        del g[t]
        del demand[t]
    supply[f] -= v
    if supply[f] == 0:
        for k, n in demand.items():
            if n != 0:
                g[k].remove(f)
        del g[f]
        del supply[f]

for n in cols:
    print("\t",n,end="")
print("")
cost = 0
for g in sorted(c_d):
    print (g,"\t",end="")
    for n in cols:
        y = res[g][n]
        if y != 0:
            print (y,end="")
        cost += y * c_d[g][n]
        print("\t",end="")
    print("")
print("Total Cost = ", cost)
```

```python
import numpy as np

def northWestCornerMethod(supply,demand,matrix):
    northwest = 0
    r = 0
    while r < len(supply):
        c = northwest
        while c < len(demand): 
            quantity = min(supply[r], demand[c])
            if quantity > 0:
                matrix[r][c] = quantity
                supply[r] -= quantity
                demand[c] -= quantity

                if supply[r] == 0: 
                    northwest = c
                    break
            c += 1
        r += 1
    return matrix

a,b = 0,0
a = int(input('Sources: '))
b = int(input('Destinations: '))
supply = []
demand = []
sum_s,sum_d = 0,0
print('Enter the supply for the sources: ')
i = 0
for i in range (0,a):
    ele = int(input())
    sum_s = sum_s + ele
    supply.append(ele)
print('Enter the demand for the destinations: ')
for i in range (0,b):
    ele = int(input())
    sum_d = sum_d + ele
    demand.append(ele)
print(supply)
print(demand)

print('Enter cost martix values row wise')
cost_matrix = [[int(input()) for x in range (a)] for y in range(b)]
print(cost_matrix)
print('Solution for NWCM: ')
matrix = np.zeros((a,b),dtype = int)

if sum_s != sum_d:
    if sum_s>sum_d:
        dummy = np.zeros(b,dtype = int)
        cost_matrix1 = np.insert(cost_matrix,a,dummy,axis = 1)
        matrix1 = np.insert(matrix,a,dummy,axis = 1)
        demand.append(sum_s-sum_d)
    else:
        dummy = np.zeros(a,dtype = int)
        cost_matrix1 = np.insert(cost_matrix,b,dummy,axis = 0)
        matrix1 = np.insert(matrix,b,dummy,axis = 0)
        supply.append(sum_d-sum_s)
else:
    cost_matrix1 = cost_matrix
    matrix1 = matrix

print(northWestCornerMethod(supply,demand,matrix1))
res = np.multiply(cost_matrix1,matrix1)
Total_cost = np.sum(res)

print('Total cost: ',Total_cost)
```
